```python
import keras
from keras.models import load_model
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
import pickle
import numpy as np
import cv2
import random
from sklearn.model_selection import train_test_split
```

```python
## DATA YUKLENIYOR

file = open(r"/content/drive/MyDrive/BioVision/YARIISMA_1/TRAINING/dataset.pkl", "rb")
data = pickle.load(file)
file.close()

inmeYok = data['inmeYok']
iskemi = data['iskemi']
kanama = data['kanama']
del data
data = None

inmeVar = np.concatenate((iskemi,kanama))
del iskemi
del kanama
iskemi = None
kananma = None

inmeYok_Y = np.zeros((len(inmeYok)))
inmeVar_Y = np.ones((len(inmeVar)))
```

```python
## DATA YUKLENIYOR

inmeVarTrain_X , inmeVarTrain_Y , inmeVarTest_X , inmeVarTest_Y = train_test_split(inmeVar,inmeVar_Y,test_size = 0.15,random_state = 42)
del inmeVar
inmeVar = None 


inmeYokTrain_X , inmeYokTrain_Y , inmeYokTest_X , inmeYokTest_Y = train_test_split(inmeYok,inmeYok_Y,test_size = 0.15,random_state = 42)
del inmeYok
inmeYok = None

print(str(len(inmeYokTrain_X)) + " " + str(len(inmeVarTrain_X)))

list_inmeYok = np.random.random_integers(len(inmeYokTrain_X)-1,size=(len(inmeVarTrain_X)))
select_no_hemoraj = []
for i in range(len(inmeVarTrain_X)):
    select_no_hemoraj.append(inmeYokTrain_X[list_inmeYok[i]])


del inmeYokTrain_X
inmeYokTrain_X = None

train_x = np.concatenate((np.array(select_no_hemoraj),inmeVarTrain_X),0)
train_y = np.concatenate((np.zeros(len(select_no_hemoraj)),np.ones(len(inmeVarTrain_X))))
del select_no_hemoraj
del inmeVarTrain_X
select_no_hemoraj = None
inmeVarTrain_X = None
```

```python
## DATA YUKLENIYOR

train_x,test_x,train_y,test_y = train_test_split(train_x,train_y,test_size = 0.3,random_state=43)

train_yy = []
for i in range(len(train_y)):
    if(train_y[i] == 0):
        train_yy.append([1,0])
    else:
        train_yy.append([0,1])
train_y = np.array(train_yy)

test_yy = []
for i in range(len(test_y)):
    if(test_y[i] == 0):
        test_yy.append([1,0])
    else:
        test_yy.append([0,1])
test_y = np.array(test_yy)
```

```python
## HER MODELIN EN IYI IKI TANESI YUKLENIYOR

path="/content/drive/MyDrive/BioVision/EN IYI MODELLER"

modelBetMobileNet = load_model(path +"/BETUL/MobileNet/ep1134batch_size32lrn_rate0_03rndm42testsize0_3acc%92.9453262786596")
modelBetMobileNet2 = load_model(path +"/BETUL/MobileNet/ep1134batch_size64lrn_rate0_01rndm42testsize0_3acc%91.88712522045856")

modelBetResNet50 = load_model(path +"/BETUL/ResNet50/ep1134batch_size32lrn_rate0_03rndm42testsize0_3acc%92.7689594356261")
modelBetResNet50_2 = load_model(path +"/BETUL/ResNet50/ep1134batch_size32lrn_rate0_03rndm42testsize0_3acc%92.50440917107584")

modelAhmVGG19 = load_model(path + "/AHMET/VGG19/ep27bt64lrn0_01rndm42tstsz0_3acc_93.03350970017637")
modelAhmVGG19_2 = load_model(path + "/AHMET/VGG19/ep32bt64lrn0_1rndm42tstsz0_3acc_92.85714285714286")

modelAhmInception = load_model(path + "/AHMET/INCEPTION/ep28bt64lrn0_01rndm42tstsz0_3acc_92.68077601410934")
modelAhmInception2 = load_model(path + "/AHMET/INCEPTION/ep39bt64lrn0_1rndm42tstsz0_3acc_92.68077601410934")

modelOnurDenseNet = load_model(path + "/ONUR/DenseNet121/Ep=30-Batch=32-LRate=_3-Rndm=43-TestSze=_30-Acc=0.9144620895385742")
modelOnurDenseNet2 = load_model(path + "/ONUR/DenseNet121/Ep=30-Batch=32-LRate=_3-Rndm=43-TestSze=_30-Acc=0.9144620895385742")

modelOnurEfficient = load_model(path + "/ONUR/EfficientNet3/Ep=30-Batch=16-LRate=_3-Rndm=43-TestSze=_30-Acc=0.9215167760848999")
modelOnurEfficient2 = load_model(path + "/ONUR/EfficientNet3/Ep=30-Batch=32-LRate=_3-Rndm=43-TestSze=_30-Acc=0.9162257313728333")
```

```python
## MODELLER TEST EDILIYOR

predictBetMobileNet = modelBetMobileNet.predict(test_x)
predictBetMobileNet2 = modelBetMobileNet2.predict(test_x)

predictBetResNet50 = modelBetResNet50.predict(test_x)
predictBetResNet50_2 = modelBetResNet50_2.predict(test_x)

predictAhmVGG19 = modelAhmVGG19.predict(test_x)
predictAhmVGG19_2 = modelAhmVGG19_2.predict(test_x)

predictAhmInception = modelAhmInception.predict(test_x)
predictAhmInception2 = modelAhmInception2.predict(test_x)

predictOnurDenseNet = modelOnurDenseNet.predict(test_x)
predictOnurDenseNet2 = modelOnurDenseNet2.predict(test_x)

predictOnurEfficient = modelOnurEfficient.predict(test_x)
predictOnurEfficient2 = modelOnurEfficient2.predict(test_x)
```

```python
## HESAPLAMA FONKSIYONLARI TANIMLANIYOR 
## calculate --> tahmin listesi dönüyor
## calculateFusion --> 6 modelin tahminlerini fusion ile işliyor

def calculate(predict,str1):
  predict_int = []
  for i in range(len(predict)):
    if(predict[i][0]>0.5):
      predict_int.append(0)
    else:
      predict_int.append(1)
  dogru = 0
  yanlis = 0
  for i in range(len(predict)):
    if(test_y[i][predict_int[i]] == 1):
      dogru += 1
    else:
      yanlis += 1
  print(str1 + " Basari : %" + str(dogru/(dogru+yanlis)*100))
  return predict_int
```

```python
## HER MODELIN SINIF TAHMINLERI ALINIYOR

predictIntBetMobileNet = calculate(predictBetMobileNet,"BetMobileNet")
predictIntBetMobileNet2 = calculate(predictBetMobileNet2,"BetMobileNet")

predictIntBetResNet50 = calculate(predictBetResNet50,"BetResNet50")
predictIntBetResNet50_2 = calculate(predictBetResNet50_2,"BetResNet50")

predictIntAhmVGG19 = calculate(predictAhmVGG19,"AhmVGG19")
predictIntAhmVGG19_2 = calculate(predictAhmVGG19_2,"AhmVGG19")

predictIntAhmInception = calculate(predictAhmInception,"AhmInception")
predictIntAhmInception2 = calculate(predictAhmInception2,"AhmInception")

predictIntOnurDenseNet = calculate(predictOnurDenseNet,"OnurDesNet")
predictIntOnurDenseNet2 = calculate(predictOnurDenseNet2,"OnurDesNet")

predictIntOnurEfficient = calculate(predictOnurEfficient,"OnurEfficient")
predictIntOnurEfficient2 = calculate(predictOnurEfficient2,"OnurEfficient")

BetMobileNet = []
BetMobileNet.append(predictIntBetMobileNet)
BetMobileNet.append(predictIntBetMobileNet2)

BetResNet50 = []
BetResNet50.append(predictIntBetResNet50)
BetResNet50.append(predictIntBetResNet50_2)

AhmetVGG19 = []
AhmetVGG19.append(predictIntAhmVGG19)
AhmetVGG19.append(predictIntAhmVGG19_2)

AhmetInception = []
AhmetInception.append(predictIntAhmInception)
AhmetInception.append(predictIntAhmInception2)

OnurDenseNet = []
OnurDenseNet.append(predictIntOnurDenseNet)
OnurDenseNet.append(predictIntOnurDenseNet2)

OnurEfficient = []
OnurEfficient.append(predictIntOnurEfficient)
OnurEfficient.append(predictIntOnurEfficient2)
```

```python
def calculateFusion(predictIntBetMobileNet,predictIntBetResNet50,predictIntAhmVGG19,predictIntAhmInception,predictIntOnurDenseNet,predictIntOnurEfficient,string):
  predictIntGeneral = []
  for i in range(len(test_y)):
    inmeYok = 0
    inmeVar = 0

    if(predictIntBetMobileNet[i] == 0):
      inmeYok += 1
    else:
      inmeVar += 1

    if(predictIntBetResNet50[i] == 0):
      inmeYok += 1
    else:
      inmeVar += 1

    if(predictIntAhmVGG19[i] == 0):
      inmeYok += 1
    else:
      inmeVar += 1

    if(predictIntAhmInception[i] == 0):
      inmeYok += 1
    else:
      inmeVar += 1

    if(predictIntOnurDenseNet[i] == 0):
      inmeYok += 1
    else:
      inmeVar += 1

    if(predictIntOnurEfficient[i] == 0):
      inmeYok += 1
    else:
      inmeVar += 1

    if(inmeYok > inmeVar):
      predictIntGeneral.append(0)
    elif(inmeYok < inmeVar):
      predictIntGeneral.append(1)
    else:
      rnd = random.randint(0,1)
      predictIntGeneral.append(rnd)
  dogru = 0
  yanlis = 0
  for i in range(len(test_y)):
      if(test_y[i][predictIntGeneral[i]] == 1):
        dogru += 1
      else:
        yanlis += 1
  print(string + "Genel Basari : %" + str(dogru/(dogru+yanlis)*100))

## EN IYI MODEL TOPLULUGU SECILIYOR

for betMobileIndex in range(2):
  for betResnet50Index in range(2):
    for ahmetVGG19Index in range(2):
      for ahmetInceptionIndex in range(2):
        for onurDenseNetIndex in range(2):
          for onurEfficientIndex in range(2):
            string = "Grup Kodu : " + str(betMobileIndex) + str(betResnet50Index) + str(ahmetVGG19Index) + str(ahmetInceptionIndex) + str(onurDenseNetIndex) + str(onurEfficientIndex) + " ---> "
            calculateFusion(BetMobileNet[betMobileIndex],BetResNet50[betResnet50Index],AhmetVGG19[ahmetVGG19Index],AhmetInception[ahmetInceptionIndex],OnurDenseNet[onurDenseNetIndex],OnurEfficient[onurEfficientIndex],string)
```
